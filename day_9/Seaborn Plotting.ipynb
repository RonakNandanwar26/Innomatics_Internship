### Univariate Box Plot
In univariate box plot we only use one column to make a boxplot.

```python
sns.boxplot(x=df['Salary'])
plt.title('Plot Name : BoxPlot')
print('Plot Name : BoxPlot')
print("Plot Type : Univariate")
print('Data Type of variable used is : ' + str(df.Salary.dtypes))
```

In above box plot, we can see that The salary ranges from 0 to 4 in Lacs. 
* Box plot generally shows us that is there any outliers available or not. 
* It decides using quantiles and $IQR(Inter \ Quantile \ range)$. 
* There are 5 verticle lines in Box Plot as shown above two are left most and right most line and 3 lines are present in box.
* First verticle line represents $ 0^{th} $ percentile or Minimum in dataset. we can represent it using formula : $ Q_1 - 1.5*IQR $
* $ 2^{nd} $ verticle line which is the part of box is $ 25^{th} $ percentile which is $ Q1 $.
* $ 3^{rd} $ line is Median or $ 50^{th} $ percentile. 
* $ 4^{th} $ line is $ 75^{th} $ percentile which is $ Q3 $.
* $ IQR = Q3 - Q1 $
* the $ 5^{th} $ line or rightmost line is Maximum or $ 100^{th} $ percentile. we can represent maximum using formula $ Q_1 + 1.5*IQR $
* The dots below Minimum and after Maximum is known as outliers. outliers are basically the values whic are more than maximum. we need to handle it using some techniques.

### KDE(Kernel Density Estimate)

```python
sns.kdeplot(data=df,x='Salary')
plt.title('no. of Persons in particular range of salary.')
print('Plot Name : Kernael Density Estimation')
print("Plot Type : Univariate")
print('Data Type of variable used is : ' + str(df.Salary.dtypes))
```

In above graph KDE, represent the data density wise.
* we can assume KDE as the line drawn on histograms.
* KDE shows that how our data is distributed.
* In above univariate KDE graph of salary, we can see that for o to 1 lacs of salary graph is ver densed and then density is decreased. 
* It shows that people having salaryrange of 0 to 1 lacs is very high as compared to another salary ranges.

### Pairplot

```python
sub_df = df[['Salary','10percentage','12percentage','CollegeTier']]
```

```python
sns.pairplot(data=sub_df,diag_kind='kde')
plt.title('Pair Plot')
print('Plot Name : Pair Plot')
print("Plot Type : Multivariate")
print('Data Type of variable used is : ' + str(df.Salary.dtypes) + ' ' + str(df['10percentage'].dtypes) + ' ' + str(df['12percentage'].dtypes) + ' ' + str(df.CollegeTier.dtypes) + ' vs ' +str(df.Salary.dtypes)+ ' ' + str(df['10percentage'].dtypes) + ' ' + str(df['12percentage'].dtypes) + ' ' + str(df.CollegeTier.dtypes) + 'All vs All')
```

As human we cannot draw graph whose dimension is more than 3. Practically it is not possible but using pairplot we can see all vs all relation graph.   
here we can see pairplot gives us the graph of all columns vs all column.   
All diagonal graphs are same column vs same column graph, which is histogram with diag_kind = 'kde', if we remove this parameter, the diagonal graphs will be histograms.

```python
sns.violinplot(data=df,x='Gender',y='Salary')
plt.title('violin plot of Gender vs Salary')
print('Plot Name : violinPlot')
print("Plot Type : Bivariate")
print('Data Type of variable used is : ' + str(df.Gender.dtypes) + ' vs ' + str(df.Salary.dtypes))
```

A violin plot is a method of plotting numeric data.   
It is similar to a box plot, with the addition of a rotated kernel density plot on each side.    
While a box plot only shows summary statistics such as mean/median and interquartile ranges, the violin plot shows the full distribution of the data.

we can observe that male candidate have somewhat higher salary then female candidate in some case.

```python
sns.displot(x='Salary',data=df,bins=25)
plt.title('Salary displot')
print('Plot Name : DisPlot')
print("Plot Type : Univariate")
print('Data Type of variable used is : ' + str(df.Salary.dtypes))
```

displot is almost same like histogram. It's kind parameter is bydefault histogram. if we want to chanege it we can make it "kde" or "ecdf". According to our need we can change it.   
If we don't provide any kind parameter then it will show us histogram.   
If we provide kde, it will show us kde plot according to data and same for ecdf plot.
The features will be according to the kind parameter.   
Here in above graph we can see that, people count with salary ranges from 0 to 0.5 lacs is very high. and people count from other salary ranges is very low.
