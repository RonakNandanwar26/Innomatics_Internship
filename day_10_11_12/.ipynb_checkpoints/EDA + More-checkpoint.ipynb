## here we can see that the numerical features have very different ranges for that we need to normalize or standardize the data, and we need to convert categorical column into numeric using some encoding techniques or dummy variables

we don't need to standardise or normalise income column because that is our dependent or output column.

```python
# ind_df is df having independeny features and dep_df is dependent column which is income in this dataset
ind_df = df.drop(['income'],axis=1)
dep_df = df.income
```

we need to standardized the numeric data in this dataset, for that we will usr the StandardScalar from sklearn. for that we need to have only numeric column in dataframe

```python
df.head()
```

```python
df.dtypes
```

```python
num_col = df.select_dtypes(include='int64')
num_col.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(num_col),columns=['age','fnlwgt','educational-num','capital-gain','capital-loss','hours-per-week'])
scaled_df.head()
```

```python
str_col = df.select_dtypes(include='object')
str_col.head()
```

now we need to convert this string type to numeric type

```python
str_col.workclass.unique()
```

```python
new_df = pd.get_dummies(df['workclass'],drop_first=True)
```

```python
new_df
```

```python
str_col.education.unique()
```

```python
edi_dummy = pd.get_dummies(df['education'],drop_first=True)
```

```python
edi_dummy
```

```python
new_df
```

```python
new_df = pd.concat([new_df,edi_dummy],axis=1)
```

```python
new_df
```

```python
str_col['marital-status'].unique()
```

we can convert not married to 0 and married to 1

```python
dct = {
    'not married':0,
    'married':1
}
str_col['marital-status'] = str_col['marital-status'].map(dct)
str_col
```

```python
new_df = pd.concat([new_df,str_col['marital-status']],axis=1)
new_df.head()
```

```python
str_col.occupation.unique()
```

```python
occu_dummy = pd.get_dummies(df['occupation'],drop_first=True)
occu_dummy
```

```python
new_df = pd.concat([new_df,occu_dummy],axis=1)
new_df
```

```python
str_col.relationship.unique()
```

```python
relation_dummy = pd.get_dummies(str_col['relationship'],drop_first=True)
relation_dummy
```

```python
new_df = pd.concat([new_df,relation_dummy],axis=1)
```

```python
new_df
```

```python
str_col.race.unique()
```

```python
race_dummy = pd.get_dummies(str_col['race'],drop_first=True)
race_dummy
```

```python
new_df = pd.concat([new_df,race_dummy],axis=1)
new_df.head()
```

```python
dct = {
    'Female':0,
    'Male':1
}
str_col['gender'] = str_col['gender'].map(dct)
str_col.head()
```

```python
new_df = pd.concat([new_df,str_col['gender']],axis=1)
new_df.head()
```

```python
str_col['native-country'].unique()
```

```python
country_dummy = pd.get_dummies(str_col['native-country'],drop_first=True)
country_dummy
```

```python
new_df = pd.concat([new_df,country_dummy],axis=1)
```

```python
new_df.head()
```

```python
dct = {
    '<=50K':0,
    '>50K':1
}
str_col.income = str_col.income.map(dct)
str_col.head()
```

```python
new_df = pd.concat([new_df,str_col['income']],axis=1)
new_df.head()
```

Now we have converted object or string type to numeric.
Now we can concat our standardized data with this new converted data.

```python
new_df = pd.concat([scaled_df,new_df],axis=1)
new_df.head()
```

Now our data is ready to go in ML or DL model for predictions.
We have independent features and dependent feature(income)
