### Mean

A mean is simply average of data.
There may be 2 types of mean.
1.) Sample Mean
2.) Population Mean

> Formula of sample mean can be given as:
$$ \overline{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $$

> Formula of population mean is given as:
$$ \mu = \frac{1}{N} \sum_{i=1}^{N} x_i $$


e.g.

data = {3,9,5,10,10,1,7,8,5,5}

$$ \overline{x} = \frac{1}{n} \sum_{i=1}^{n} x_i $$

$$ \overline{x} = \frac{3+9+5+10+10+1+7+8+5+5}{10}  $$

$$ \overline{x} = \frac{63}{10}  $$

$$ \overline{x} = 6.3  $$

```python
print('Average Monthly income is ' + str(df['Mthly_HH_Income'].mean()))
```

### Median

Median is a middle value of data after arranging the data in ascending order.

$$

        \frac{\left( \frac{n}{2} \right)^{th} element + \left( \frac{n}{2} + 1 \right)^{th} element}{2}    & \mbox{if } \ n \ is \ even \\
    \end{array}
\right.
$$

e.g.

data = {3,5,5,1,7,5,10,9,8,10}

arranged_data_in_ascending = {1,3,5,5,5,7,8,9,10,10}

here n is 10(even)

so, according to the above formula

$$ median = \frac{\left( 5+7 \right)}{2} $$

### Covariance

If we want to find out the relation between two features from the dataset, covariance is used. covariance helps us to quantify the relationship between features, between random variables in particular dataset.

|size of house||Price|
|---||---|
|1200 sqm||100k\$|
|1500 sqm||200k\$|
|1800 sqm||300k\$|


> As the size of house increases -> Price increases

$$ cov(size,price) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu_x)*(y_i- \mu_y) $$

if $ x \uparrow y \uparrow \longrightarrow cov +ve $ \
if $ x \uparrow y \downarrow \longrightarrow cov -ve $ \
if $ x \downarrow y \downarrow \longrightarrow cov +ve $


We can relate covariance and variance with each other.

$$ variance \ \sigma^2(x) = \frac{\sum_{i=1}^{N} \left( x_i - \mu \right)^2}{N}  $$

$$ variance \ \sigma^2(x) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \mu_x) * (x_i - \mu_x) $$

$$ cov(x,x) = var(x) $$

```python
df['Mthly_HH_Expense'].cov(df['Mthly_HH_Income'])
```

#### Spearman's rank correlation coefficient

Pearson gives correlation assuming that data is linear.

Spearman rank correlation coefficient  gives correlation even if data is non-linear.

```python
Image(filename='spearman.png')
```

##### Formula

$$ r_s = \rho_{rg_x,rg_y} = \frac{cov(rg_x,rg_y)}{\sigma_{rg_x}\sigma_{rg_y}} $$
where $\rho$ denotes the pearson corelation coefficient but applied to rank variables. \
$ cov(rg_x,rg_y) $ is the covariance of the rank variables. \
$ \sigma_{rg_x}  \sigma_{rg_y} $ are the standard deviation of the rank variables.

Another formula to find coefficient is 
$$ r_s = 1 - \frac{6\sum d_{i}^{2}}{n(n^2 -1)} $$

$ d_i  = rg(x_i) - rg(y_i)$ is the difference between two rank of each observation. \
$ n $ is no. of observation.


For Example:

The raw data below is used to show the correlation between the IQ of person with the no. of hours spent infront of tv per week.


```python
Image(filename='spearman_example.jpeg')
```

#### Formula to find spearman rank correlation coefficient is
$$ r_s = 1 - \frac{6\sum d_{i}^{2}}{n(n^2 -1)} $$

here, firstly we need to evaluate $ d_i^2 $. To do so we need to follow below steps.
1.) sort the data by the first column $ x_i $. create new column $ x_i $ and assign it the ranked values from $ 1,2,3...,n $ \
2.) Next, sort the data by the second column $ (y_i) $, create a fourth column $ y_i $ and similarly assign it the ranked values $ 1,2,3...,n $ \
3.)create a fifth column $ d_i $ to hold the difference between two rank column $ (x_i \&  y_i) $ \
4.) create one final column, $ d_i^2 $ to hold the value of column $ d_i $ squared.

These four steps are done in next image.

```python
Image(filename='spearman_ex_solution.jpeg')
```

Now we need to calculate spearman rank correlation coefficient.

$$ r_s = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} $$ \
$$ r_s = 1 - \frac{6*194}{10(100 - 1)} $$ \
$$ r_s = -0.17575 $$ \
which shows that relation between IQ and no. of hours spent on tv is very less.

### Normal Distribution/Gaussian Distribution

If a dataset is equally distributed across the plane or dispersed equally across the mean, in that case we are able to plot a bell shaped curve as shown below.

```python
Image(filename='bell_curve.png')
```

suppose we have dataset D = {1,2,3,4,1,2,100,110}, here dataset is dispersed between 1 and 110.

In this case we will get graph something like below. It is known as left skewness. we can't pass data like this in ML or DL model, we need to normalize the data and get bell shaped curve.

```python
Image(filename='left_skewed.jpeg')
```

For getting bell shaped curve from this data we can apply any scaling technique on data. \
e.g. applying square root on data. \
If we apply sqare root on above data then \
$ \sqrt{1} = 1 $  \
$ \sqrt{110} = 10.48 $ \
Now our data ranges between 1 to 10.48, using this data we can get somewhat bellshaped curve. \
we can apply log too for scaling \
If we apply log then \
$ log_{10}(1) = 0 $ \
$ log_{10}(110) = 2.04 $ \
here we get our data ranges from 0 to 2, which will give us more good bell shaped curve. 

Here, we are just trying to change the scale of data, we are not changing the actual value of data or actual significance of data. 

here, There is one more thing that, we have only 1 dimensional data and we are plotting 2 dimensional graph without having 2d data. 

The answer for this is, we can find value of y using the given data using below equation. 

$$ y = \frac{1}{\sigma \sqrt{2 \pi}} e^{\frac{-1}{2} (\frac{x - \mu}{\sigma})^2}, -\infty \leq x \leq +\infty $$

we can check the curve is bess shaped or left skewed or right skewed using above formula using given dataset.

when we normalize the data $ \mu $ could be anything and $ \sigma $ could be anything.

#### Features of Normal Distribution

* The mean, mode and median are all equal.
* The curve is symmetric at the center (i.e. around the mean, ).
* Exactly half of the values are to the left of center and exactly half the values are to the right.
* The total area under the curve is 1.

### Effect on Mean, Median and Mode due to Skewness

* if the distribution of data is skewed to the left, the mean is less than the median, which is often less than the mode. 
* If the distribution of data is skewed to the right, the mode is often less than the median, which is less than the mean.
