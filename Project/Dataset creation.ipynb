```python
import pandas as pd
import os
import shutil
```

```python
# data creation for covid19 positive
Images_Path = 'covid_chestxray/images'
```

```python
df = pd.read_csv('covid_chestxray/metadata.csv')
df
```

```python
df.shape
```

```python
Target_dir = 'Datasets/Covid+'
if not os.path.exists(Target_dir):
    os.mkdir(Target_dir)
    print('Covid Folder is created')
```

```python
count=0
for i,row in df.iterrows():
    if 'COVID-19' in row['finding'] and row['folder']=='images':
        filename = row['filename']
        image_path = os.path.join(Images_Path,filename)
        image_copy_path = os.path.join(Target_dir,filename)
        shutil.copy2(image_path,image_copy_path)
        count+=1
```

```python
print(count)
```

```python
Kaggle_File_Path = 'chest_xray_kaggle/train/NORMAL'
Target_Normal_Path = 'Datasets/Normal'
```

```python
# kaggle dataset is too big, so making 50% normal and 50% covid+ data, if we want whole data just copy whole normal data from kaggle folder 
image_names = os.listdir(Kaggle_File_Path)
image_names
```

```python
import random # for randomly selecting normal images from all normal chest xrays
random.shuffle(image_names)

```

```python
for i in range(563):
    image_name = image_names[i]
    image_path = os.path.join(Kaggle_File_Path,image_name)
    target_path = os.path.join(Target_Normal_Path,image_name)
    
    shutil.copy2(image_path,target_path)
    print('Copying image ',i)
    
```

```python
# Making Train, Validation and test dataset

normal_image_path = 'Datasets/Normal'
covid_image_path = 'Datasets/Covid+'
```

```python
# For Normal Images
normal_images = os.listdir(normal_image_path)
normal_images
```

```python
random.shuffle(normal_images)
```

```python
for i in range(563):
    if i<=563*0.75:  
        image_name = normal_images[i]
        image_path = os.path.join(normal_image_path,image_name)
        target_path = os.path.join('New_Dataset/train/normal',image_name)

        shutil.copy2(image_path,target_path)
        print('Copying normal image in train ',i)
    else:
        image_name = normal_images[i]
        image_path = os.path.join(normal_image_path,image_name)
        target_path = os.path.join('New_Dataset/test/normal',image_name)

        shutil.copy2(image_path,target_path)
        print('Copying normal image in test ',i)
```

```python
# Now for covid images
covid_images = os.listdir(covid_image_path)
covid_images
```

```python
random.shuffle(covid_images)
```

```python
for i in range(563):
    if i<=563*0.75:  
        image_name = covid_images[i]
        image_path = os.path.join(covid_image_path,image_name)
        target_path = os.path.join('New_Dataset/train/covid+',image_name)

        shutil.copy2(image_path,target_path)
        print('Copying covid image in train ',i)
    else:
        image_name = covid_images[i]
        image_path = os.path.join(covid_image_path,image_name)
        target_path = os.path.join('New_Dataset/test/covid+',image_name)

        shutil.copy2(image_path,target_path)
        print('Copying covid image in test ',i)
```
